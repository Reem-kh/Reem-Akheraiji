```python
# Initial interaction with user:
print()
Name = input("Hello, what’s your name?")
print()
LongName = input( Name + ", I understand you’ve been going to college. What’s the name of your school?")
print()
ShortName = input( LongName + "... that’s a long name. Does it go by a shorter name?")

print()
if ShortName == "Yes" :
    UniName = input("What's your university's shorter name?")
else :
    UniName = LongName
    
print()
print("Very good, " + Name + ".  Let’s talk about your grades from last quarter at " + UniName + ".")
print()
print("I need to know the classes you took, how much each is worth in terms of credits, and the grade you made. When you’ve entered all your classes and their information, just hit enter at the prompt for class designation and number.")
print()   
input("Sound good?")
print()   

#Lists & Dictionary:

ClassList = []
CreditList = []
GradeList = []
list_of_lists = [ClassList, CreditList, GradeList]
Points = {"A":4,"B":3,"C":2,"D":0,"F":-1}
PointsList = []

#Course information entry and data validation:

CourseIsBad = True
while CourseIsBad :
    C = (input("Please tell me the designation and number of a class you took last quarter at " + UniName + "."))
    if C == "" :
        CourseIsBad = False 
    if CourseIsBad == True :
        print()  
        ClassList.append(C)
        CreditisBad = True
        while CreditisBad :
            CR = int(input("How many credits is " + C + " worth?"))
            CreditList.append(CR)
            if CR in range (1,7) :
                CreditisBad = False
                GradeisBad = True
                while GradeisBad :
                    print() 
                    GR = input("And what grade did you make in " + C + "?")
                    GradeList.append(GR)
                    if GR == "A" or GR == "B" or GR == "C" or GR == "D" or GR == "F":
                        print()   
                        print("Next Class")
                        print()   
                        GradeisBad = False
                    else :
                        print("Sorry you must enter one of the letter grades A, B, C, D, or F")  
            else :
                print("Sorry you must enter a credits number from 1 to 6")

                
#Total number of classes and credits
CourseCount = len(ClassList)

CreditCount = 0 
for x in range (0, len(CreditList)) :
    CreditCount = CreditCount + CreditList[x]
    
print()                     

print("Okay, " + Name + ", you took " + str(CourseCount) + " classes last quarter for a total of " + str(CreditCount) + " credits at " + UniName + 
". In sorted order by designation,         they were:")

#Headings
HClass   = ('\x1B[4m' + str("CLASS") + '\x1B[0m')
HCredits = ('\x1B[4m' + str("NUMBER OF CREDITS") + '\x1B[0m')
HGrades  = ('\x1B[4m' + str("GRADE") + '\x1B[0m')
HPoints  = ('\x1B[4m' + str("POINTS") + '\x1B[0m')

# List of classes in sorted order by designation:

print()   
print(HClass + "     " + HCredits + "  " + HGrades)

for x in range (0, len(ClassList)-1) :
    for y in range(x + 1, len(ClassList)) :
        if ClassList[y] < ClassList[x] :
            ClassList[y], ClassList[x] = ClassList[x], ClassList[y]
            CreditList[y], CreditList[x] = CreditList[x], CreditList[y]
            GradeList[y], GradeList[x] = GradeList[x], GradeList[y]
            
for x in range (0, len(ClassList)) :
    print(ClassList[x],"        ", str(CreditList[x]),"         ",  str(GradeList[x]))

     
print()   

# The class with the highest grade:

BestGrade = min(GradeList)
GradeLocation = GradeList.index(min(GradeList))
ClassName = ClassList[GradeLocation]
CreditLocation = GradeList.index(min(GradeList))
CreditsNumber = CreditList[CreditLocation]

print("Your best class was " + ClassName + " in which you made an " + BestGrade + " for the " + str(CreditsNumber) + " credits.") 

# The class with the lowest grade:

WorstGrade = max(GradeList)
GradeLocation = GradeList.index(max(GradeList))
ClassName = ClassList[GradeLocation]
CreditLocation = GradeList.index(max(GradeList))
CreditsNumber = CreditList[CreditLocation]

print("Your worst class was " + ClassName + " in which you made an " + WorstGrade + " for the " + str(CreditsNumber) + " credits.") 

print()   

# GPA Calculations:

TotalCredits = 0 
for x in range (0, len(CreditList)) :
    TotalCredits = TotalCredits + CreditList[x]
    
for x in range (0, len(ClassList)):
    PT = CreditList[x]*Points[GradeList[x]]
    PointsList.append(PT)

TotalPoints = sum(PointsList)

GPA = format(TotalPoints/TotalCredits ,",.2f")

print() 

print("Your GPA for last quarter was " + str(GPA) + " at " + UniName + ".")

print() 
print() 

print("From best to worst, here are your classes for last quarter:")

print() 

# List of classes in sorted order by grades:

print(HClass + "       " + HGrades)

for x in range (0, len(GradeList)-1) :
    for y in range(x + 1, len(GradeList)) :
        if GradeList[y] < GradeList[x] :
            GradeList[y], GradeList[x] = GradeList[x], GradeList[y]
            ClassList[y], ClassList[x] = ClassList[x], ClassList[y]
                 
for x in range (0, len(GradeList)) :
    print(ClassList[x],"   ",  str(GradeList[x]))

print()   

# List of classes in sorted order by designation:

print()   

print("This is how I computed your GPA:")

print()   
print(HClass + "     " + HCredits + "  " + HGrades + "  " + HPoints)

for x in range (0, len(ClassList)-1):
    for y in range(x + 1, len(ClassList)) :
        if ClassList[y] < ClassList[x] :
            ClassList[y], ClassList[x]   = ClassList[x], ClassList[y]
            GradeList[y], GradeList[x]   = GradeList[x], GradeList[y]
            
for x in range (0, len(ClassList)-1):
    for y in range(x + 1, len(ClassList)) :
        if ClassList[y] < ClassList[x] :
            ClassList[y], ClassList[x]   = ClassList[x], ClassList[y]
            CreditList[y], CreditList[x]   = CreditList[x], CreditList[y]

for x in range (0, len(ClassList)-1):
    for y in range(x + 1, len(ClassList)) :
        if ClassList[y] < ClassList[x] :
            ClassList[y], ClassList[x]   = ClassList[x], ClassList[y]
            PointsList[y], PointsList[x]   = PointsList[x], CreditList[y]

for x in range (0, len(ClassList)) :
    print(ClassList[x],"       ", str(CreditList[x]),"          ",  str(GradeList[x]),"    ", str(PointsList[x]))

# How GPA was computed:

print()   

print("Total Points: " + str(TotalPoints))

print()   

print("Total Credits: " + str(TotalCredits))

print()   

print(str(TotalPoints) + " divided by " + str(TotalCredits) + " is equal to a " + GPA + " " + "GPA.")
```
